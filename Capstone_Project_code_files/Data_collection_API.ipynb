```python
for item in data :
  Player_Statistics = item['statistics']

  for player_index in Player_Statistics :

    team_name_list.append(player_index['team']['name'])

    league_list.append(player_index['league']['name'])

    appearances_list.append(player_index['games']['appearences'])
    lineups_list.append(player_index['games']['lineups'])
    minutes_list.append(player_index['games']['minutes'])
    number_list.append(player_index['games']['number'])
    position_list.append(player_index['games']['position'])
    rating_list.append(player_index['games']['rating'])
    captain_list.append(player_index['games']['captain'])

    in_list.append(player_index['substitutes']['in'])
    out_list.append(player_index['substitutes']['out'])
    bench_list.append(player_index['substitutes']['in'])

    total_list.append(player_index['shots']['total'])
    on_list.append(player_index['shots']['on'])

    total_goals_list.append(player_index['goals']['total'])
    conceded_list.append(player_index['goals']['conceded'])
    assists_list.append(player_index['goals']['assists'])
    saves_list.append(player_index['goals']['saves'])

    total_passes_list.append(player_index['passes']['total'])
    key_passes_list.append(player_index['passes']['key'])
    accuracy_passes_list.append(player_index['passes']['accuracy'])

    total_tackles_list.append(player_index['tackles']['total'])
    blocks_list.append(player_index['tackles']['blocks'])
    interceptions_list.append(player_index['tackles']['interceptions'])

    total_duels_list.append(player_index['duels']['total'])
    duels_won_list.append(player_index['duels']['won'])

    attempts_list.append(player_index['dribbles']['attempts'])
    success_list.append(player_index['dribbles']['success'])
    past_list.append(player_index['dribbles']['past'])

    drawn_list.append(player_index['fouls']['drawn'])
    committed_list.append(player_index['fouls']['committed'])

    yellow_list.append(player_index['cards']['yellow'])
    red_list.append(player_index['cards']['red'])

    won_list.append(player_index['penalty']['won'])
    commited_list.append(player_index['penalty']['commited'])
    scored_list.append(player_index['penalty']['scored'])
    missed_list.append(player_index['penalty']['missed'])
    saved_list.append(player_index['penalty']['saved'])
```
