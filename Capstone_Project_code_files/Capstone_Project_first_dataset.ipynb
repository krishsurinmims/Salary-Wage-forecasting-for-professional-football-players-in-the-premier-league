```python
Initial_capstone_dataset = ('/content/drive/My Drive/Colab Notebooks/Capstone_Project/Inital_capstone_dataset.csv')
```

```python
first_df = pd.read_csv(Initial_capstone_dataset)

first_df.shape
```

```python
first_df.head()
```

```python
#Filling all the NaN values with 0
first_df.fillna(0, inplace=True)
```

```python
#Checking to see if there are any missing values left
print(first_df.isnull().sum())
```

```python
first_df.info()
```

```python
# Remove rows where 'Minutes' is 0
first_df = first_df[first_df['Minutes'] != 0]
```

```python
#Checking the shape of the data after removing the players who have not played
first_df.shape
```

Based on extensive research, players with fewer than 300 minutes in the season often provide limited information and can introduce noise into the data. Such sparse data points can adversely affect the model's performance by creating biases. To ensure a robust and reliable model, we are excluding players with less than 300 minutes.

```python
first_df = first_df[first_df['Minutes'] > 300]
```
