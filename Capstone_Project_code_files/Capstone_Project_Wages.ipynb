Arsenal

```python
response = requests.get(url="https://www.spotrac.com/epl/arsenal-fc/payroll/2022/")
html_doc = response.text

soup = BeautifulSoup(html_doc, "html.parser")
```

```python
Wages_Arsenal = soup.find_all(name="span", class_="cap info")
Wages_list_Arsenal = []
for wage in Wages_Arsenal :
  wage_text = wage.getText()
  Wages_list_Arsenal.append(wage_text)
```

```python
final_arsenal_wages = Wages_list_Arsenal[::2]
print(final_arsenal_wages)
```

```python
Arsenal_Players = soup.select("tbody tr td a")
Arsenal_player_names = [player.getText() for player in Arsenal_Players]
print(Arsenal_player_names)
```

```python
# create a list of tuples using enumerate()
tuples = [(key, value)
          for i, (key, value) in enumerate(zip(Arsenal_player_names, final_arsenal_wages))]

# convert list of tuples to dictionary using dict()
arsenal_dict = dict(tuples)
print(arsenal_dict)
```

Now we need to combine all the dictionaries to obtain one final wages dictionary with the player names as the keys and their wages as values

```python
merged_dict = dict(arsenal_dict.items() | AV_dict.items() | BHA_dict.items() | brentford_dict.items() | bournemouth_dict.items() | Chelsea_dict.items() | CP_dict.items() | ManC_dict.items() | ManU_dict.items() | fulham_dict.items() | everton_dict.items() | southampton_dict.items() | leicester_dict.items() | Wolves_dict.items() | WH_dict.items() | Spurs_dict.items() | NF_dict.items() | NU_dict.items() | liverpool_dict.items())
```

```python
print(merged_dict)
```

```python
# Creating a full name column to match the keys in the merged dict
data['Full Name'] = data['First Name'] + ' ' + data['Last Name']
```

```python
# Creating the wages column using the values in the merged dict
data['Wages'] = data['Full Name'].map(merged_dict)
```
